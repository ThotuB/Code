# Trial version

```python
from random import random
```

```python
def roll(p: float, clovers: int) -> bool:
    if random() < p:
        return True
    
    if clovers <= 0:
        return False
    
    return roll(p, clovers - 1)

def trial(p: float, clovers: int, n: int) -> float:
    pos = 0
    
    for _ in range(n):
        if roll(p, clovers):
            pos += 1
            
    return pos / n
```

```python
from numpy import arange
```

```python
step = 0.01
trials = 10000

x = arange(0, 1 + step, step)
```

```python
ys = [[trial(prob, clovers, trials) for prob in x] for clovers in range(5)]
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()
for num_clovers, y in enumerate(ys):
    fig.add_trace(go.Scatter(x=x, y=y, name=f"Clovers: {num_clovers}"))
    
fig.update_layout(
    hovermode="x unified",
)
fig.show()
```

# Math version

```python
def luck_probability(p: float, clovers: int) -> float:
    n = 1 - p
    return 1 - n ** (clovers + 1) 
```

```python
ys = [[luck_probability(prob, clovers) for prob in x] for clovers in range(5)]
```

```python
fig = go.Figure()
for num_clovers, y in enumerate(ys):
    fig.add_trace(go.Scatter(x=x, y=y, name=f"Clovers: {num_clovers}"))
    
fig.update_layout(
    hovermode="x unified",
)
fig.show()
```
